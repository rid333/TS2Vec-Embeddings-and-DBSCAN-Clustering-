```python
import pandas as pd
from ts2vec import TS2Vec
import numpy as np
import torch
from sklearn.decomposition import PCA
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.manifold import TSNE
import mplcursors

plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Arial'
```

```python
df = pd.read_csv("data_main.csv")
```

```python
df.head()
```

```python
scaler_flex = StandardScaler()
flex_scaled = scaler_flex.fit_transform(df[["flex_right", "flex_left"]])
```

```python
pca = PCA(n_components=1)
df["flex_pca"] = pca.fit_transform(flex_scaled)
```

```python
scaler_all = StandardScaler()
X_scaled = scaler_all.fit_transform(df[["heart_rate", "flex_pca"]])
```

```python
scaled_df = pd.DataFrame(X_scaled, columns=["heart_rate_scaled", "flex_pca_scaled"])
scaled_df["Time"] = df["Time"]

# Plot
plt.figure(figsize=(10,6))
plt.plot(scaled_df["Time"], scaled_df["heart_rate_scaled"], label="Heart Rate (scaled)")
plt.plot(scaled_df["Time"], scaled_df["flex_pca_scaled"], label="Flex PCA (scaled)")

plt.xlabel("Time (Minute)")
plt.ylabel("Scaled Value")
plt.legend()

step_plot = 60
plt.xticks(scaled_df["Time"].iloc[::step_plot], rotation=45)

plt.tight_layout()
plt.show()
```

```python
def sliding_windows(data, window_size=10, step=5):
    """
    data: (T, D)
    returns: (N, W, D) windows
    """
    T, D = data.shape
    windows = []
    for start in range(0, T - window_size + 1, step):
        end = start + window_size
        windows.append(data[start:end])
    return np.array(windows)
```

```python
window_size = 30
step = 5

windows = sliding_windows(X_scaled, window_size, step)
```

```python
windows.shape
```

```python
# Tell the model to expect 2 features per time step
model = TS2Vec(input_dims=2, output_dims=32)
model.fit(windows, n_epochs=20, verbose=True)
embeddings = model.encode(windows, encoding_window="full_series")
print(f"Embeddings shape: {embeddings.shape}")
```

```python
# Function to plot k-distance graph
def plot_k_distance_graph(X, k):
    neigh = NearestNeighbors(n_neighbors=k)
    neigh.fit(X)
    distances, _ = neigh.kneighbors(X)
    distances = np.sort(distances[:, k-1])
    plt.figure(figsize=(10, 6))
    plt.plot(distances)
    plt.xlabel('Points')
    plt.ylabel(f'{k}-th nearest neighbor distance')
    plt.grid()
    plt.show()
# Plot k-distance graph
plot_k_distance_graph(embeddings, k=5)
```

```python
dbscan = DBSCAN(eps=1.5, min_samples=16)
cluster_labels = dbscan.fit_predict(embeddings)
print(f"Unique labels found: {np.unique(cluster_labels)}")
```

```python
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
tsne_results = tsne.fit_transform(embeddings)
print(f"t-SNE results shape: {tsne_results.shape}")  # Should be (150, 2)
```

```python
%matplotlib widget
# Code for displaying data on hover in the plot
# Training weights were not saved to ensure results remain consistent with those reported in the article.
# You can retrain the model to reproduce similar results.
fig, ax = plt.subplots(figsize=(10, 8))

scatter = ax.scatter(
    tsne_results[:, 0],
    tsne_results[:, 1],
    c=cluster_labels,  # colored by DBSCAN cluster
    cmap="viridis",
    alpha=0.8,
)

ax.set_title("t-SNE of TS2Vec Embeddings (DBSCAN Clusters)")
ax.set_xlabel("t-SNE 1")
ax.set_ylabel("t-SNE 2")

# --- Add interactive cursor ---
cursor = mplcursors.cursor(scatter, hover=True)


@cursor.connect("add")
def on_hover(sel):
    idx = sel.index
    start = idx * step
    end = start + window_size
    raw_window = df.iloc[start:end][["heart_rate", "flex_right", "flex_left"]]
    means = raw_window.mean().round(2).to_dict()
    vars_ = raw_window.var().round(2).to_dict()
    if "Time" in df.columns:
        t_start = df.iloc[start]["Time"]
        t_end = df.iloc[end - 1]["Time"]
        time_text = f"Time: {t_start} → {t_end}"
    else:
        time_text = f"Index range: {start}–{end}"

    text = (
        f"Idx: {idx}\nCluster: {cluster_labels[idx]}\n"
        f"{time_text}\n"
        f"heart_rate μ={means['heart_rate']}, σ²={vars_['heart_rate']}\n"
        f"flex_right μ={means['flex_right']}, σ²={vars_['flex_right']}\n"
        f"flex_left μ={means['flex_left']}, σ²={vars_['flex_left']}"
    )

    sel.annotation.set_text(text)
    sel.annotation.get_bbox_patch().set(fc="black", alpha=0.7)


plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

scatter = ax.scatter(
    tsne_results[:, 0],
    tsne_results[:, 1],
    c=cluster_labels,
    cmap="viridis",
    alpha=0.8,
)

uniq, counts = np.unique(cluster_labels, return_counts=True)
handles = []
for lab, cnt in zip(uniq, counts):
    color = scatter.cmap(scatter.norm(lab))
    name = "Noise" if lab == -1 else f"Cluster {lab}"
    handles.append(
        Line2D([0], [0],
               marker='o', linestyle='none',
               markerfacecolor=color, markeredgecolor='none',
               markersize=8, label=f"{name} (n={cnt})")
    )

ax.legend(handles=handles, frameon=False)

ax.set_xlabel("t-SNE 1")
ax.set_ylabel("t-SNE 2")

plt.show()
```

```python
signals = ["heart_rate", "flex_right", "flex_left"]

rows = []
n_windows = len(cluster_labels)
for idx in range(n_windows):
    start = idx * step
    end = start + window_size
    raw_window = df.iloc[start:end][signals]

    row = {"cluster": int(cluster_labels[idx])}
    for s in signals:
        row[f"{s}_mean"] = raw_window[s].mean()
        row[f"{s}_var"]  = raw_window[s].var()
    rows.append(row)

window_stats = pd.DataFrame(rows)

cluster_stats = (
    window_stats.groupby("cluster")
    .agg("mean")
    .join(window_stats.groupby("cluster").size().rename("n_windows"))
    .sort_index()
    .round(3)
)

print(cluster_stats)
```
